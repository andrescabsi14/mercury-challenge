This notebook shows the distribution of metrics for the 100 runs of the baserate model for MA over the scoring period.

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use("fivethirtyeight")
import os
import re
import json
```

```python
def plot_metric(result_dict, cc_name, met_name, ax):
    _values = cc_dict[met_name]
    if met_name != "Lead Time":
        pd.Series(_values).plot(kind="density", ax=ax, xlim=(0,1), label="Density")
    else:
        pd.Series(_values).plot(kind="density", ax=ax, label="Density")
    _mean = np.mean(_values)
    _std = np.std(_values)
    _thresh = _mean + .5*_std
    ax.axvline(x=_mean, label="Mean = {:.4f}".format(_mean))
    ax.axvline(x=_thresh, color="darkgreen", linestyle="dashed", 
               label="Threshold = {:.4f}".format(_thresh))
    ax.set_title(met_name)
    ax.legend()
    return ax
    
    
```

```python
MBR_FILENAME = 'MBR Metrics Details 2018-08-07 to 2018-10-12.json'
```

```python
MC_HOME = os.path.join("..", "..", "..")
DATA_HOME = os.path.join(MC_HOME, "data")
MA_SR_PATH = os.path.join(DATA_HOME, "scoring_results", "MA")
MBR_FILEPATH = os.path.join(MA_SR_PATH, MBR_FILENAME)
EVT_COUNTRIES = ["Egypt", "Iraq", "Jordan", "Lebanon", "Saudi Arabia", "Syria"]

with open(MBR_FILEPATH, "r") as f:
    mbr = json.load(f)
```

The values of the metrics for each run can be found in the directory `data/scoring_results/MA`

## Egypt

The graphs below show the kernel density estimates for the distribution of the Mercury Score, Quality Score, F1, and Lead Time metric based on the 100 values provided from the repeated runs of the Baserate model.  The solid vertical line shows the mean value and the dashed line shows the "Baserate + 0.5 standard deviations" threshold.  For countries with relatively few events there will be some spread to the distribution.  For very eventful countries there will be little spread.  Please note that the spread for Lead Time is to be taken with a grain of salt, as the Baserate model simulates submission time using a Poisson distribution for the expected number of days before the predicted `Event_Date` the warning is to be submitted.

```python
cc = "Egypt"
cc_dict = mbr[cc]

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
ms_ax = axes[0,0]
qs_ax = axes[0,1]
f1_ax = axes[1,0]
lt_ax = axes[1,1]
ms_ax = plot_metric(mbr, cc, "Mercury Score", ms_ax)
qs_ax = plot_metric(mbr, cc, "Quality Score", qs_ax)
f1_ax = plot_metric(mbr, cc, "F1", f1_ax)
lt_ax = plot_metric(mbr, cc, "Lead Time", lt_ax)
fig.suptitle("Metrics Distribution for {}".format(cc), size=20, y=1.02)
fig.tight_layout()
```
